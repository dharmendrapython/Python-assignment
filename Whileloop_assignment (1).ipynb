```python
import random  
number = random.randint(1,100)  
print (number) 
```

```python
Attempts = 0
guess_number = int(input("Enter a number between 1 to 100: " )) 
while guess_number != number:
    Attempts += 1
    if guess_number < number:  
        print ("you are below the number, try again")
        guess_number = int(input("Enter a number between 1 to 100: " ))
    elif guess_number <= 40:
        print ("you are  to close, try again")
        guess_number = int(input("Enter a number between 1 to 100: " ))
    elif guess_number <= 50:
         print ("you are close, try again")
         guess_number = int(input("Enter a number between 1 to 100: " ))
    elif guess_number <= 60:
        print ("you are far, try again")
        guess_number = int(input("Enter a number between 1 to 100: " ))
    elif guess_number > 60:  
        print ("you are so far, try again")
        guess_number = int(input("Enter a number between 1 to 100: " ))
else:
        Attempts += 1
        print("you guess the correct number, congratulations") 
        print ("Total Attempts:", Attempts)
```
